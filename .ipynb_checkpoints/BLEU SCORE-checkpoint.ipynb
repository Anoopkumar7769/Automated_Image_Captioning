```python
from nltk.translate.bleu_score import sentence_bleu
reference = [
    'A couple watches a boat against a skyline'.split(),
    'A man and woman sit on a bench  watching a boat go by'.split(),
    'The sun is setting while a man and woman watch a boat go by'.split(),
    'Two people sit on a bench and watch a boat on the water'.split(),
    'Two people watching a boat sail past'.split()
]
candidate = 'Two people sit on a bench '.split()
print('BLEU score -> {}'.format(sentence_bleu(reference, candidate )))
```
